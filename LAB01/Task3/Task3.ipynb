```python
# bubble sort algorithm given in the question.
def bubbleSort(test_arr):
  for i in range(arr_len-1):
    for j in range(arr_len-i-1): 
      if test_arr[j] > test_arr[j+1]:
        test_arr[j], test_arr[j+1] = test_arr[j+1], test_arr[j]
```

```python
arr_len = int(input3_testCase2.readline().strip())
#print(arr_len)
arr = input3_testCase2.readline().split()
#print(arr)
test_arr = [0]*arr_len
for i in range(arr_len):
  test_arr[i] = int(arr[i])
#print(test_arr)

# testing if the array is already sorted in the first place.
for i in range(arr_len-1):                      # since the loop runs for n times so it should be θ(n).
  if test_arr[i] < test_arr[i+1]:
  #   print("Yes!")
  # else:
  #   print("NO")
    flag = "sorted!"
  else:
    flag = "NOT sorted"
    break
#print(flag)

'''
This "if" block will only turn out to be True if the array is  NOT SORTED. In that case, bubble sort algorithm will be 
implemented on the array giving us a sorted array. But if the "if" block happens to be false then this means that the 
array is sorted in the first place. So, it will get in the else block and the bubble sort algo will not get implemnted. 
The code from line 34-37 will simply write out the actual array in the txt file.  
'''
if flag == "NOT sorted":
  bubbleSort(test_arr)          
else:
  pass
  
# writing in output file
space = " "
for i in range(arr_len):
  output3.write(str(test_arr[i]))
  output3.write(space)
```

```python
input3_testCase2.close()
output3.close()
'''
Thus, in the best case i.e. where all the elements of the array is sorted, only one loop will run (line#11) 
till n giving us time complexity of θ(n) overtaking the whole bubble sort algo.
'''
```
