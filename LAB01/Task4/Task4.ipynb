```python
from re import I

N = int(input4_testCase1.readline().strip())
#print(N)

all = (input4_testCase1.readlines())
#print(all)

# Making a list of Student ID's.
id = all[0].split()
#print(id)
Si = [0]*N
#print(Si)
idx_Si = 0
for i in id:
  Si[idx_Si] = int(i)
  idx_Si += 1
#print(Si)

# making a list of student marks.
marks = all[1].split()
#print(marks)
Sm = [0]*N
idx_Sm = 0
for i in marks:
  Sm[idx_Sm] = int(i)
  idx_Sm += 1
#print(Sm)

# Applying SELECTIOIN SORT as it takes min no of swaps as required by the  q.
for i in range(len(Sm)):
  for j in range(i):
    if Sm[j]<Sm[i]:
      # swapping marks array
      temp_Sm = Sm[j]
      Sm[j] = Sm[i]
      Sm[i] = temp_Sm
      # swapping ID array
      temp_Si = Si[j]
      Si[j] = Si[i]
      Si[i] = temp_Si
#print(Si, Sm)

# Sorting ID array when more than one individual have the same marks.
min = 0
for i in range(N):
  for i in range(N-1):
    if Sm[i] == Sm[i+1]:
      if Si[i] > Si[i+1]:
        Temp = Si[i]
        Si[i] = Si[i+1]
        Si[i+1] = Temp
#print(Si)

# writing answer in text file.
for i in range(N):
  ans = f"ID: {Si[i]} Mark: {Sm[i]} \n"
  output4.write(ans)
```
