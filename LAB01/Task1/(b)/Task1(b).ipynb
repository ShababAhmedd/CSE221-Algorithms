```python
def calculator(eqn):

  if "+" in eqn:
    x = eqn.split("+")
    y = int(x[0])+int(x[1])

  elif "-" in eqn:
    x = eqn.split("-")
    y = int(x[0])-int(x[1])

  elif "*" in eqn:
    x = eqn.split("*")
    y = int(x[0])*int(x[1])  
  
  elif "/" in eqn:
    x = eqn.split("/")
    y = int(x[0])/int(x[1])

  return y

# print(calculator("67 + 41"))
# print(calculator("85 / 5"))
# print(calculator("13 - 56"))
# print(calculator("12 * 19"))
```

```python
t = int(input1b.readline().strip())
#print(t)

for i in range(t):
  line1 = input1b.readline().strip()       # how is strip exactly working?
  #print(line)
  line2 = line1[10:]
  #print(line2)
  numeric = calculator(line2)
  #print(numeric)
  output1b.write(f"The result of {line2} is {numeric} \n")
```
