```python
# insertion sort function
def insertion_sort(list1):  

  for i in range(1, len(list1)):  
  
    value = list1[i]  
    j = i - 1  
    while j >= 0 and value > list1[j]:  
      list1[j + 1] = list1[j]  
      j -= 1  
      list1[j + 1] = value  
  return list1  
        
# list1 = [1,5,0,8,4,3]
# x = insertion_sort(list1)
# print(x)
```

```python
N = int(input1_testCase1.readline().strip())
#print("N :", N)

id = input1_testCase1.readline().strip().split()
#print("ID :", id)

m = input1_testCase1.readline().strip().split()
#print("m : ", m)
marks = [0]*N
for i in range(N):
  marks[i] = int(m[i])
#print("Marks: ", marks)

# applying insertin sort on marks as well as swapping the ID's according to it..
for i in range(1, N):  
  
    value = marks[i] 
    value1 = id[i] 
    j = i - 1  
    while j >= 0 and value > marks[j]:  
      marks[j + 1] = marks[j] 
      id[j + 1] = id[j] 
      j -= 1  
      marks[j + 1] = value
      id[j + 1] = value1 
#print(marks)
#print(id)  
```

```python
# writing in file and finally closeing both the files.
space = " "
for i in id:
  output1.write(i)
  output1.write(space)

input1_testCase1.close()
output1.close()
```
